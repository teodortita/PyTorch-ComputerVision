# VGG Main

```python
# Load Data MNIST
batch_size = 100
train_loader, test_loader, valid_loader = load_data('MNIST', batch_size)
```

```python
# Create Model
model = VGG(arch='VGG16', 
            channels=1, 
            num_classes=10)

print(model)
print("No of parmeter:",sum(p.numel() for p in model.parameters() if p.requires_grad))
```

```python
# Train Model With Early Stopping
train(model,
      train_loader, 
      valid_loader, 
      epochs=50,
      name='vgg16_mnist.pt')
```

```python
# Test Model
test(model, 
     test_loader, 
     name='vgg16_mnist.pt',
     batch_size = batch_size)
```
